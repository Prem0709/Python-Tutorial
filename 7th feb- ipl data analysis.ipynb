```python
plt.plot(Salary[0], c= "g", ls= "-.", marker= "o", ms= 7)
plt.xticks(list(range(0,10)), Seasons)
plt.show()
```

1. **Salary[0]** â†’ Extracts the first row of the Salary matrix, which contains Sachin's salary over 10 seasons.
2. **c="g"** â†’ Sets the line color to green (g).
3. **ls="-."** â†’ Uses a dash-dot (-.) line style.
4. **marker="o"** â†’ Places circle (o) markers at each data point.
5. **ms=7** â†’ Sets the marker size to 7.

```python
plt.plot(Salary[0], c= "b", ls= "-.", marker= "s", ms= 10,label= Players[0])
plt.xticks(list(range(0,10)), Seasons,rotation= "vertical")
plt.show()
```

- **`Salary[0]`** â†’ Extracts the **first row** of the `Salary` matrix, which represents **Sachin's salary over 10 seasons**.  
- **`c='Green'`** â†’ Sets the **line color** to **green** (**`c="g"`** also works).  
- **`ls=':'`** â†’ Uses a **dotted (`:`) line style**.  
- **`marker='s'`** â†’ Uses **square (`s`) markers** for each data point.  
- **`ms=7`** â†’ Sets the **marker size** to **7**.  
- **`label=Players[0]`** â†’ Uses the **first playerâ€™s name** (**"Sachin"**) as a **legend label**.
